## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
from matplotlib.pyplot import ylabel
# gráfico do exercício 1

diamonds = data[['cut', 'price']]

diamonds = data.query('price > 17500')


with sns.axes_style('darkgrid'):

  outlier_graf = sns.histplot(data=diamonds, x='price')
  outlier_graf.set(title='Preços Outliers', xlabel='Preços', ylabel='Contagem')
```

**Insight do gráfico 1**: A quantidade mais expressiva preços ouliers está entre o valor de 18200 a 18300.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

diamonds = data[['carat','clarity','price']]

with sns.axes_style('darkgrid'):

  cor_graf = sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', palette='pastel')
  cor_graf.set(title='Relação entre Preço x Peso', xlabel='Peso', ylabel='Preço')
  cor_graf.get_legend().set_title("Tranparência")
  cor_graf.figure.set_size_inches(w=40/2.54, h=20/2.54)


  
```

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

with sns.axes_style('whitegrid'):

  cars_graf = sns.boxplot(x=data["weight"])
  cars_graf.set(title='Distribuição de peso', xlabel='Peso');
```

**Insight do gráfico 3**: A maioria dos carros tem um peso apróximado entre 2800 a 3600.
